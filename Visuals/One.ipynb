```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set(style="whitegrid")
```

```python
data = pd.read_csv('C:\\Users\\sidda\\OneDrive\\Desktop\\SIH v2\\Set\\Base-v2.csv')
print(data.head())
```

```python
success_rates = data.groupby(['RECEIVER REGION', 'TIME SLOT'])['DELIVERY OUTCOME'].mean().reset_index()
success_rates.columns = ['Region', 'Time Slot', 'Success Rate']
```

```python
plt.figure(figsize=(10, 6))
regions = success_rates['Region'].unique()
colors = sns.color_palette("husl", len(regions))

for region in regions:
    region_data = success_rates[success_rates['Region'] == region]
    plt.pie(region_data['Success Rate'], labels=region_data['Time Slot'], autopct='%1.1f%%', colors=colors, startangle=140)
    plt.title(f'Success Rate by Time Slot in {region}')
    plt.show()
```

```python
# Create a line plot for success rates by time slot and region
plt.figure(figsize=(12, 8))
sns.lineplot(data=success_rates, x='Time Slot', y='Success Rate', hue='Region', marker='o')
plt.title('Success Rate by Time Slot and Region')
plt.xlabel('Time Slot')
plt.ylabel('Success Rate')
plt.legend(title='Region')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

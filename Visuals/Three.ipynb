```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:\\Users\\sidda\\OneDrive\\Desktop\\SIH v2\\Set\\Base-v2.csv')

grouped_data = data.groupby(['RECEIVER REGION', 'DAY', 'DELIVERY OUTCOME']).size().unstack(fill_value=0).reset_index()

grouped_data.columns = ['Region', 'Day', 'Failures', 'Successes']

grouped_data['Total Deliveries'] = grouped_data['Failures'] + grouped_data['Successes']
```

```python
plt.figure(figsize=(14, 8))

sns.barplot(data=grouped_data, x='Region', y='Total Deliveries', hue='Day', palette='viridis', edgecolor='black')

plt.title('Region-wise Delivery Days and Successes')
plt.xlabel('Region')
plt.ylabel('Number of Deliveries')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
